```python
from donut import DonutModel
from PIL import Image
import torch

# Change the path here:
model = DonutModel.from_pretrained('/content/donut/result/train_w9/20250601_173056')
if torch.cuda.is_available():
    model.half()
    device = torch.device("cuda")
    model.to(device)
else:
    model.encoder.to(torch.bfloat16)

model.eval()

image = Image.open("/content/sample_data/b709d8e2-3865-5291-ab4d-a8a96e1901ce_w9_0.png").convert("RGB") # Change here
with torch.no_grad():
  output = model.inference(image=image, prompt="<s_w9_cord_v9>")
```
